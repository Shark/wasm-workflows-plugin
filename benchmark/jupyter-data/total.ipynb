```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from matplotlib import rcParams
```

```python
total_wasm_distributed_1 = pd.read_csv('one-at-a-time/total_wasm-distributed_1.csv', sep=';').query('workflow_result=="succeeded" and total_time_taken > 0 and total_time_taken < 100')
total_wasm_distributed_10 = pd.read_csv('one-at-a-time/total_wasm-distributed_10.csv', sep=';').query('workflow_result=="succeeded" and total_time_taken > 0 and total_time_taken < 100')
total_wasm_distributed_100 = pd.read_csv('one-at-a-time/total_wasm-distributed_100.csv', sep=';').query('workflow_result=="succeeded" and total_time_taken > 0 and total_time_taken < 100')

total_wasm_local_1 = pd.read_csv('one-at-a-time/total_wasm-local_1.csv', sep=';').query('workflow_result=="succeeded" and total_time_taken > 0 and total_time_taken < 100')
total_wasm_local_10 = pd.read_csv('one-at-a-time/total_wasm-local_10.csv', sep=';').query('workflow_result=="succeeded" and total_time_taken > 0 and total_time_taken < 100')
total_wasm_local_100 = pd.read_csv('one-at-a-time/total_wasm-local_100.csv', sep=';').query('workflow_result=="succeeded" and total_time_taken > 0 and total_time_taken < 100')

total_container_1 = pd.read_csv('one-at-a-time/total_container_1.csv', sep=';').query('workflow_result=="succeeded" and total_time_taken > 0 and total_time_taken < 100')
total_container_10 = pd.read_csv('one-at-a-time/total_container_10.csv', sep=';').query('workflow_result=="succeeded" and total_time_taken > 0 and total_time_taken < 100')
total_container_100 = pd.read_csv('one-at-a-time/total_container_100.csv', sep=';').query('workflow_result=="succeeded" and total_time_taken > 0 and total_time_taken < 100')
```

```python
data_1 = pd.DataFrame({
    "Container": total_container_1['total_time_taken'],
    "Distributed": total_wasm_distributed_1['total_time_taken'],
    "Local": total_wasm_local_1['total_time_taken'],    
})

data_10 = pd.DataFrame({
    "Container": total_container_10['total_time_taken'],
    "Distributed": total_wasm_distributed_10['total_time_taken'],
    "Local": total_wasm_local_10['total_time_taken'],
})

data_100 = pd.DataFrame({
    "Container": total_container_100['total_time_taken'],
    "Distributed": total_wasm_distributed_100['total_time_taken'],
    "Local": total_wasm_local_100['total_time_taken']
})

sns.set_style("whitegrid")
fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
#fig.suptitle("Workflow Runtime")
axs[0].set_title("1 parallel workflow step")
axs[0].set_ylabel("seconds")
axs[1].set_title("10 parallel workflow steps")
axs[2].set_title("100 parallel workflow steps")
g = sns.stripplot(ax=axs[0], data=data_1)
g = sns.stripplot(ax=axs[1], data=data_10)
g = sns.stripplot(ax=axs[2], data=data_100)
```
